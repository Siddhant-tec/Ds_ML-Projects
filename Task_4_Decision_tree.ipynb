# Task #4 Decision Tree  

# Importing Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from sklearn import tree
```

# Loading Datasets

```python
data = datasets.load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.DataFrame({"Target":data.target})

pd.concat([df,y], axis = 1).head(10)
```

# Creating a Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(df,y)
```

# Plotting Tree

```python
plt.figure(figsize = (6, 4), dpi=300)
tree.plot_tree(dtree, feature_names = data.feature_names, class_names = data.target_names, filled = True, rounded = True);
plt.title("Iris Classification Tree")
plt.show()
```
